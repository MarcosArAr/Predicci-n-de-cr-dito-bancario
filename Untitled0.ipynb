#Preprocesamiento de la data

```python
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(columns=["Personal Loan"])
y=df["Personal Loan"]
```

```python
X.info()
```

```python
num_selector = make_column_selector(dtype_include='number')

standard_scaler = StandardScaler()

imputer_num = SimpleImputer(strategy="mean")
```

```python
col_transformer = make_column_transformer((make_pipeline(imputer_num,standard_scaler), num_selector),
                        remainder = 'passthrough')

col_transformer
```

```python
col_transformer.fit(X)
```

```python
aux=col_transformer.transform(X)
X_procesado=pd.DataFrame(aux,columns=col_transformer.get_feature_names_out())
X_procesado.info()
X_procesado
```

```python
X_train, X_test, y_train, y_test=train_test_split(X_procesado,y,random_state=30,test_size=0.3)
```

#EvaluaciÃ³n de PCA

Utilizando el PCA con un umbral de varianza de 95%, nos muestra que el modelo debe seleccionar el nÃºmero de componentes principales necesarios para explicar al menos el 95% de la varianza total en los datos originales.

image.png

En el grafico, vemos que con un componente 11 obtenemos una varianza cercana a 1, es decir una varianza similar a si no reducimos el nÃºmero de componentes. En conclusiÃ³n se seleccionarÃ¡ el nÃºmero de componentes principales que en conjunto capturan el 95% de la informaciÃ³n contenida en las variables originales que para este caso seran 11.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
pca=PCA(0.95)
pca.fit(X_procesado)
X_pca=pca.transform(X_procesado)
X_pca=pd.DataFrame(X_pca,columns=pca.get_feature_names_out())
X_pca.info()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.plot(range(1,X_pca.shape[1]+1),pca.explained_variance_ratio_.cumsum(),marker='o')
plt.title('Curva de PCA')
plt.xlabel('NÃºmero de componentes principales')
plt.ylabel('Varianza explicada acumulada')
plt.axvline(11, color='k', linestyle='--')
plt.grid()
plt.show()
```

## Decision Tree

```python
dt=DecisionTreeClassifier(max_depth=4)
dt.fit(X_train, y_train)
print(dt.score(X_train,y_train))
print(dt.score(X_test,y_test))
y_pred_dt=dt.predict(X_test)
print(classification_report(y_test, y_pred_dt))

# Genera la matriz de confusiÃ³n
cm = confusion_matrix(y_test, y_pred_dt)

# Crea una figura y un conjunto de subtramas
fig, ax = plt.subplots()

# Crea un mapa de calor a partir de la matriz de confusiÃ³n
sns.heatmap(cm, annot=True, fmt='d', ax=ax, cmap="BuGn", cbar=True)

# AÃ±ade etiquetas, tÃ­tulo y elimina las marcas de verificaciÃ³n
ax.set_xlabel('Predicho')
ax.set_ylabel('Verdadero')
ax.set_title('Matriz de ConfusiÃ³n')
ax.xaxis.set_ticklabels(['No aceptÃ³', 'AceptÃ³'])
ax.yaxis.set_ticklabels(['No aceptÃ³', 'AceptÃ³'])

plt.show()
```
