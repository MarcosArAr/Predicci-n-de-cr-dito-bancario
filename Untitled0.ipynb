#Preprocesamiento de la data

```python
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(columns=["Personal Loan"])
y=df["Personal Loan"]
```

```python
X.info()
```

```python
num_selector = make_column_selector(dtype_include='number')

standard_scaler = StandardScaler()

imputer_num = SimpleImputer(strategy="mean")
```

```python
col_transformer = make_column_transformer((make_pipeline(imputer_num,standard_scaler), num_selector),
                        remainder = 'passthrough')

col_transformer
```

```python
col_transformer.fit(X)
```

```python
aux=col_transformer.transform(X)
X_procesado=pd.DataFrame(aux,columns=col_transformer.get_feature_names_out())
X_procesado.info()
X_procesado
```

```python
X_train, X_test, y_train, y_test=train_test_split(X_procesado,y,random_state=30,test_size=0.3)
```

#Evaluación de PCA

Utilizando el PCA con un umbral de varianza de 95%, nos muestra que el modelo debe seleccionar el número de componentes principales necesarios para explicar al menos el 95% de la varianza total en los datos originales.

image.png

En el grafico, vemos que con un componente 11 obtenemos una varianza cercana a 1, es decir una varianza similar a si no reducimos el número de componentes. En conclusión se seleccionará el número de componentes principales que en conjunto capturan el 95% de la información contenida en las variables originales que para este caso seran 11.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
pca=PCA(0.95)
pca.fit(X_procesado)
X_pca=pca.transform(X_procesado)
X_pca=pd.DataFrame(X_pca,columns=pca.get_feature_names_out())
X_pca.info()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.plot(range(1,X_pca.shape[1]+1),pca.explained_variance_ratio_.cumsum(),marker='o')
plt.title('Curva de PCA')
plt.xlabel('Número de componentes principales')
plt.ylabel('Varianza explicada acumulada')
plt.axvline(11, color='k', linestyle='--')
plt.grid()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

## Decision Tree
Se obtuvo:


*   Score Train 0.9931408973992569


*   Score Test 0.9873333333333333



```python
dt=DecisionTreeClassifier(max_depth=4)
dt.fit(X_train, y_train)
print(dt.score(X_train,y_train))
print(dt.score(X_test,y_test))
y_pred_dt=dt.predict(X_test)
print(classification_report(y_test, y_pred_dt))

# Genera la matriz de confusión
cm = confusion_matrix(y_test, y_pred_dt)

# Crea una figura y un conjunto de subtramas
fig, ax = plt.subplots()

# Crea un mapa de calor a partir de la matriz de confusión
sns.heatmap(cm, annot=True, fmt='d', ax=ax, cmap="BuGn", cbar=True)

# Añade etiquetas, título y elimina las marcas de verificación
ax.set_xlabel('Predicho')
ax.set_ylabel('Verdadero')
ax.set_title('Matriz de Confusión')
ax.xaxis.set_ticklabels(['No aceptó', 'Aceptó'])
ax.yaxis.set_ticklabels(['No aceptó', 'Aceptó'])

plt.show()
```

```python
accuracy = []
for x in range(1, 30):
  clf = DecisionTreeClassifier(max_depth=x, random_state=0)
  clf.fit(X_train, y_train)
  score = clf.score(X_test, y_test)
  accuracy.append(score)
```

```python
accuracy
```

```python
plt.plot(range(1, 30), accuracy)
plt.ylabel('Accuracy')
plt.xlabel('Depth')
plt.title('Relación entre la precisión y la profundidad')
plt.show()
```

En el grafico se observa que a partir de un "Depth" de 3 obtenemos un accuracy mayor al 0.98. Asi mismo se observa que con un depth de 6 se obtendrá un mayor accuracy respecto a los demás valores de depth.

```python
dt=DecisionTreeClassifier(max_depth=7)
dt.fit(X_train, y_train)
print(dt.score(X_train,y_train))
print(dt.score(X_test,y_test))
y_pred_dt=dt.predict(X_test)
print(classification_report(y_test, y_pred_dt))

# Genera la matriz de confusión
cm = confusion_matrix(y_test, y_pred_dt)

# Crea una figura y un conjunto de subtramas
fig, ax = plt.subplots()

# Crea un mapa de calor a partir de la matriz de confusión
sns.heatmap(cm, annot=True, fmt='d', ax=ax, cmap="BuGn", cbar=True)

# Añade etiquetas, título y elimina las marcas de verificación
ax.set_xlabel('Predicho')
ax.set_ylabel('Verdadero')
ax.set_title('Matriz de Confusión')
ax.xaxis.set_ticklabels(['No aceptó', 'Aceptó'])
ax.yaxis.set_ticklabels(['No aceptó', 'Aceptó'])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Escalamiento
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Input shape
input_shape = X_train.shape[1]
print(input_shape)
```

## Modelo 1
Se obtuvo:



*   Loss: 0.10954036563634872
*   Accuracy: 0.9639999866485596







```python
model = Sequential()
model.add(Dense(13,
                input_dim = input_shape,
                activation = 'relu'))
model.add(Dense(6,
                activation = 'relu'))
model.add(Dense(1, activation = "sigmoid"))
model.summary()
```

```python
model.compile(loss = 'bce', optimizer = 'adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train,
                    validation_data = (X_test, y_test),
                    epochs=10)
```

```python
plt.plot(history.history['loss'], label='Train loss', color = "blue")
plt.plot(history.history['val_loss'], label='Validation Loss', color="red")
plt.title("Pérdida (loss) durante el entrenamiento y la validación del modelo")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend();
```

En el gráfico de pérdida durante el entrenamiento y validación del modelo observamos que conforme aumenta el número de épocas, el valor de loss empieza a disminuir. Esto aplica tanto para el valor de pérdida de train y validation.
Asi mismo vemos que el comportamiento de train y de validation entre las epocas 2 y 6 es similar, sin embargo fuera de estos valores el comportamiento es discordante entre ambos.

```python
plt.plot(history.history['accuracy'], label='Train Accuracy', color = "blue")
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color = "red")
plt.title("Precisión (accuracy) durante el entrenamiento y la validación del modelo")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend();
```

En el gráfico de precisión durante el entrenamiento y validación del modelo observamos que conforme aumenta el número de épocas, el valor de accuracy empieza a aumentar. Esto aplica tanto para el valor de precisión de train y validation.
Asi mismo vemos que el comportamiento de train y de validation a partir de la época 2 es similar entre ambas.

```python
loss, acc = model.evaluate(X_test, y_test,
                            batch_size=32)
print("Loss:", loss) # Función de pérdida
print("Accuracy:", acc)
```
